```python
import os
from dotenv import load_dotenv
from openai import OpenAI

# Load environment variables from .env
load_dotenv()
api_key = os.getenv("OPENAI_API_KEY")
if not api_key:
    raise ValueError("OPENAI_API_KEY not found. Add it to your .env file.")

# OpenAI client
client = OpenAI(api_key=api_key)

# System prompt
SYSTEM_PROMPT = (
    "You are a clinical decision-support assistant for education only. "
    "You are not a doctor. Do not diagnose. Provide risk-aware, structured guidance."
)

def analyze_symptoms(symptoms: str) -> str:
    user_prompt = f"""Patient-reported symptoms: {symptoms}

Provide:
1) Possible concerns
2) Urgency level
3) Safe next steps
4) Red flags requiring emergency care"""

    response = client.chat.completions.create(
        model="gpt-4o-mini",
        temperature=0.2,
        messages=[
            {"role": "system", "content": SYSTEM_PROMPT},
            {"role": "user", "content": user_prompt},
        ],
    )
    return response.choices[0].message.content

# Example run
if __name__ == "__main__":
    symptoms = "fever, sore throat, fatigue for 3 days"
    result = analyze_symptoms(symptoms)
    print(result)
```
